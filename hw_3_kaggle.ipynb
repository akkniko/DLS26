# Метрика

Перед решением любой задачи важно понимать, как будет оцениваться ваше решение. В данном случае мы используем стандартную для задачи классификации метрику ROC-AUC. Ее можно вычислить, используя только предсказанные вероятности и истинные классы без конкретного порога классификации + она раотает даже если классы в данных сильно несбалансированны (примеров одного класса в десятки раз больше примеров длугого). Именно поэтому она очень удобна для соревнований.

Посчитать ее легко:


```python
from sklearn.metrics import roc_auc_score

y_true = [
    0,
    1,
    1,
    0,
    1
]

y_predictions = [
    0.1,
    0.9,
    0.4,
    0.6,
    0.61
]

roc_auc_score(y_true, y_predictions)
```

```python
from sklearn.linear_model import LogisticRegression, LogisticRegressionCV
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, RobustScaler, LabelEncoder, OneHotEncoder
from sklearn.pipeline import make_pipeline
```

```python
" " in data[num_cols]
```

```python
# YOUR CODE
model = LogisticRegression(class_weight='balanced')
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer

#нормировка числовых данных:
scaler = StandardScaler()
imputer = SimpleImputer(strategy='median')

numericTransf = Pipeline(steps=[
    ('scaler', StandardScaler()),
    ('SImputer', imputer)
])


#обработка категориальных:
categoricalTransf = Pipeline(steps = [
    ('ohe_encoder', OneHotEncoder(handle_unknown="ignore"))
])

#содеиняю:
preprocessor = ColumnTransformer(transformers = [
                                                ("numeric", numericTransf, num_cols),
                                                ("categorical", categoricalTransf, cat_cols)
                                                ]
                                 )


classifier = Pipeline(steps = [
    ('pip', preprocessor),
    ('clf', LogisticRegression(class_weight='balanced'))
])
classifier
```

```python
# YOUR CODE
from sklearn.model_selection import train_test_split
target_col = 'Churn'
X = data.drop(columns=[target_col]).copy()
y = data[target_col]

#ОЧИСТКА ДО РАЗДЕЛЕНИЯ (безопасно, т.к. не использует статистику)
X[num_cols] = X[num_cols].apply(lambda c: pd.to_numeric(c.astype(str).str.strip(), errors='coerce'))
X_train, X_val, y_train, y_val = train_test_split(X,y, train_size = 0.8, random_state=123)
```

```python
X_train.isnull().sum()
```

```python
# classifier.fit(X_train, y_train)

from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score,accuracy_score
classifier.fit(X_train, y_train)

y_score = classifier.score(X_val, y_val)
print(y_score)#0.747
```

```python
from catboost import CatBoostClassifier

param_grid = {
    'clf__C': [0.0001, 0.001, 0.01, 0.1, 1],
    'clf__penalty': ['l2'],                     # тип регуляризации
    'clf__max_iter': [100]
}
ctbst = CatBoostClassifier(iterations=20, learning_rate=0.001, depth=16)
grid = GridSearchCV(classifier, param_grid, cv = 5)
grid.fit(X_train, y_train)
print("Лучшие параметры:", grid.best_params_)
print("Лучший score:", f"{grid.best_score_:.4f}")
```
